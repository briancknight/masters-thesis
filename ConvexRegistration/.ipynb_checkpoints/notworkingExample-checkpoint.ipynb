```python
# A working example:

def upScaleRegistration(targetName, baseName):
    # wow this worked??
    base = readImage(baseName)
    target = readImage(targetName)
    (m, n, _) = base.copy().shape
    
    k = 4
    
    baseDownScale = readImage(baseName, (int(np.floor(n/k)), int(np.floor(m/k))))
    targetDownScale = readImage(baseName, (int(np.floor(n/k)), int(np.floor(m/k))))
    
    p = Taylor08(targetDownScale, baseDownScale, 'secondOrder')
    
    base = secondOrderDeformImage(base, -p[:6], -p[6:])
    plt.imshow(base)
    plt.show()
    
    k = 2
    
    baseDownScale = readImage(baseName, (int(np.floor(n/k)), int(np.floor(m/k))))
    targetDownScale = readImage(targetName, (int(np.floor(n/k)), int(np.floor(m/k))))
    
    p = Taylor08(targetDownScale, baseDownScale, 'secondOrder')
    
    registeredIm = secondOrderDeformImage(base, -p[:6], -p[6:])
    plt.imshow(base)
    
    return (p, registeredIm)
```

```python
upScaleRegistration('images/konaTransform.png', 'images/konaResize.png')
```
